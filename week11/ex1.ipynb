```python
#---------------------------------------Step 1.2---------------------------------------------
import numpy as np
from scipy.stats import poisson, norm

genome_size = 1_000_000    # 1 Mbp
coverage = 3      # 3× sequencing
readlength = 100     # read length
num_reads = (genome_size * coverage) // readlength

genome_coverage = np.zeros(genome_size, dtype=int)

rng = np.random.default_rng()
for _ in range(num_reads):
    startpos = rng.integers(0, genome_size - readlength + 1)
    genome_coverage[startpos:startpos + readlength] += 1

np.savetxt("coverage.txt", genome_coverage)

maxcoverage = int(genome_coverage.max())
xs = np.arange(0, maxcoverage + 1)

poisson_estimates = poisson.pmf(xs, mu=coverage)
normal_estimates = norm.pdf(xs, loc=coverage, scale=np.sqrt(coverage))

np.savetxt("coverage_values.txt", xs)
np.savetxt("poisson_estimates.txt", poisson_estimates)
np.savetxt("normal_estimates.txt", normal_estimates)
```

```python
reads = ['ATTCA', 'ATTGA', 'CATTG', 'CTTAT', 'GATTG', 'TATTT', 'TCATT', 'TCTTA', 'TGATT', 'TTATT', 'TTCAT', 'TTCTT', 'TTGAT']
k = 3

graph = set()

for read in reads:
    for i in range(len(read) - k):
        kmer1 = read[i:i+k]
        kmer2 = read[i+1:i+1+k]
        edge = f"{kmer1} -> {kmer2}"
        graph.add(edge)

with open("debruijn_graph.dot", "w") as f:
    f.write("digraph G {\n")
    for edge in sorted(graph):
        f.write(f"    {edge};\n")
    f.write("}\n")
```
